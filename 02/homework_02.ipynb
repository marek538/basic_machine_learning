````python
# dostane dataset kde se jména států shodují - vrátí true pokud je v datech v příznaku column víc než půlka nan
def more_than_half_nulls(data, column):
    nulls = 0
    for i in range(data.shape[0]):
        if pd.isnull(data.iloc[i][column]):
            nulls += 1
    
    # bugfixing
    if nulls-2 > data.shape[0]/2:
        return True
    return False
    
    
def fill_nan_by_state_mean(mean_value, dataframe, state, column):
    mask = dataframe['Country'] == state

    dataframe.loc[mask, column] = dataframe.loc[mask, column].fillna(mean_value)
    return dataframe
  
    
# v datasetu nahradí všechny nan příznaky ve sloupci column u státu state
# k výpočtu používá trénovací data, což může trochu zhoršit čitelnost kódu
def substitue_nan(data, column, state):    
#     print(column, state)
    matching_rows = data[data["Country"] == state]
    
#     print("matching_rows = ", matching_rows.shape[0])
    if not matching_rows.empty:
        first_row = matching_rows.iloc[0]
    else:
        return data
    
    mean_value = None
    if more_than_half_nulls(Xtrain.loc[Xtrain["Country"] == state], column):
        if column == "Alcohol":
            print("here")
            mean_value = Xtrain[column].median()
        elif column == "Hepatitis B":
            mean_value = Xtrain.loc[Xtrain["infant deaths"] == first_row["infant deaths"]][column].median()
        elif column == "Polio":
            mean_value = Xtrain.loc[Xtrain["infant deaths"] == first_row["infant deaths"]][column].median()
        elif column == "BMI":
            mean_value = Xtrain.loc[Xtrain["Status"] == first_row["Status"]][column].median()
        elif column == "Total expenditure":
            mean_value = Xtrain.loc[Xtrain["Status"] == first_row["Status"]][column].median()
        elif column == "Diphtheria":
            mean_value = Xtrain.loc[Xtrain["infant deaths"] == first_row["infant deaths"]][column].median()
        elif column == "GDP":
            mean_value = Xtrain.loc[Xtrain["Status"] == first_row["Status"]][column].median()
        elif column == "Population":
            mean_value = Xtrain[column].median()
        elif column == "thinness  1-19 years":
            mean_value = Xtrain.loc[Xtrain["Status"] == first_row["Status"]][column].median()
        elif column == "thinness 5-9 years":
            mean_value = Xtrain.loc[Xtrain["Status"] == first_row["Status"]][column].median()
        elif column == "Income composition of resources":
            mean_value = Xtrain.loc[(Xtrain["Schooling"] == first_row["Schooling"]) & (Xtrain["Status"] == first_row["Status"])][column].median()
        elif column == "Schooling":
            mean_value = Xtrain.loc[Xtrain["percentage expenditure"] == first_row["percentage expenditure"]][column].median() 
        
    else:
        mean_value = Xtrain.loc[Xtrain["Country"] == state][column].median()
        

    data = fill_nan_by_state_mean(mean_value, data, state, column)
#     print("mean_value = ", mean_value)
    return data


def fill_mean(data):
    nan_cols = [i for i in data.columns if data[i].isnull().any()]
   
    for col in nan_cols:
        mean_v = Xtrain[col].median()
        data[col] = data[col].fillna(mean_v)
    return data


# dolní všechny nan
def substitue_all_nan(data):
    nan_cols = [i for i in data.columns if data[i].isnull().any()]
#     print("naser si", nan_cols)
    for col in nan_cols:
#         print(data[col])
        null_in_row = data[data[col].isnull()]
#         print(null_in_row)
        states = null_in_row["Country"].unique()
#         print(states)
        for state in states:
#             print(col, state)
            data = substitue_nan(data, col, state)

    # pro jistotu, aby nezustal nikde nan
    data = fill_mean(data)
    return data

# za všechny nany dosadí -1 = mohlo by být užitečné pro les, použiju jako hyperparametr
def fill_negative(data):
    data = data.fillna(-1)
    return data
        
````

Pro velké množství unikátních hodnot příznaku country jsem se rozhodl je rozdělit do kategorií, pro které jsem vytvořil dummy příznaky. Poloha země na kontinentu by mohla hrát významnou roli pro věk dožití.

````python
# přidá pro každý kontinent dummy příznak
def add_continent_dummies(data):
    return data.assign(asia=0, europe=0, africa=0, australia_oceania=0, north_america=0, south_america=0)
    
# rozdělí země do kontinentů (ty mam vytvořené výše v polích)  
def assign_country(data):
    for index, row_data in data.iterrows():
        if row_data["Country"] in asia:
            data.at[index, "asia"] = 1
        elif row_data["Country"] in europe:
            data.at[index, "europe"] = 1
        elif row_data["Country"] in africa:
            data.at[index, "africa"] = 1
        elif row_data["Country"] in australia_oceania:
            data.at[index, "australia_oceania"] = 1
        elif row_data["Country"] in north_america:
            data.at[index, "north_america"] = 1
        elif row_data["Country"] in south_america:
            data.at[index, "south_america"] = 1
    return data
        
def drop_country(data):
    data = data.drop('Country', axis=1)
    return data

def country_to_int(data):
    data = add_continent_dummies(data)
    data = assign_country(data)
    data = drop_country(data)
    return data
    
````

````python
def transform_data(data):
    data = convert_status(data)
    data = country_to_int(data)
    return data
````

### Moje impementace random forest

````python
# Váš kód zde


########################################################
# Předpřipravená kostra modelu náhodného lesa
class CustomRandomForest:
    trees = []

    # vytvořím prázdné stromy s jejich příslušnými parametry a uložím si je do pole
    def __init__(self, n_estimators, max_samples, max_depth, **kwargs):

        self.n_estimators = n_estimators
        self.max_samples = max_samples
        self.max_depth = max_depth
        self.kwargs = kwargs
        
        for i in range(self.n_estimators):
               self.trees.append(DecisionTreeRegressor(max_depth=self.max_depth, **kwargs)) 
        
    # vygeneruje n náhoných čísel mezi 0 až Xtrain velikost, kde n je velikost * max_samples parametr   
    def bootstrap(self, size) -> list:
        result = []
        values = size * self.max_samples
        for num in range(int(values)):
            result.append(randrange(size))
        
        return result
            
    
    # natrénuje každý strom
    def fit(self, X, y):
        if isinstance(X, pd.DataFrame):
            X = X.values
 
        for tree in self.trees:
            positions = self.bootstrap(X.shape[0])
            
            xtmp = X[positions]
            ytmp = y.iloc[positions]
            
            tree.fit(xtmp, ytmp)
            


    def predict_val(self, X) -> int:
        result = []

        if isinstance(X, pd.DataFrame):
            X = X.values  # Convert DataFrame to NumPy array

        # Ensure that the shape is (1, -1) for a single sample with any number of features
        X = X.reshape(1, -1)

        for tree in self.trees:
            prediction = tree.predict(X)
            result.append(prediction)

        res = np.mean(result)

        return res
    
        
    def predict(self, X):
        if isinstance(X, pd.DataFrame):
            X = X.values
        
        # vytvořím pole, do kterého uložím výsledek
        y_predicted = np.zeros((X.shape[0],))
        
        for num in range(X.shape[0]):
            tmp = self.predict_val(X[num, :])

            y_predicted[num] = tmp
        
        return y_predicted
````

### Hřebenová regrese 
lineární jsem usoudil že nebude efektivní, kvůli závislosti příznaků

````python
lam = 1
intercept = False
best_rmse = 10000000
# (0-no scaler, 1-minmax, 2-standard)
best_data = 0

all_data = [current_data_train, X_train_minmax, X_train_std]
all_val = [current_data_val, X_val_minmax, X_val_std]

for i in range(0, 35):
    inter = True
    for j in range(2):
        for dat in range(3):
            clf_ridge = Ridge(alpha=i, fit_intercept=inter)
            clf_ridge.fit(all_data[dat], Ytrain)
            RMSE_val_ridge = mean_squared_error(Yval, clf_ridge.predict(all_val[dat]), squared = False)
            
            if RMSE_val_ridge < best_rmse:
                best_rmse = RMSE_val_ridge
                lam = i
                intercept = inter
                best_data = dat
            inter = False
````

### Náhodný les

````python
estimators = 0
data_portion = 0
best_depth = 0

best_rmse = 10000000

# (0-no scaler, 1-minmax, 2-standard)
best_data = 0

all_data = [current_data_train, X_train_minmax, X_train_std]
all_val = [current_data_val , X_val_minmax, X_val_std]

for est in range(135, 150, 5):
    for depth in range(17, 21):
        for samples in np.arange(0.4, 0.6, 0.2):
            for dat in range(3):
                forest = CustomRandomForest(n_estimators=est, max_samples=samples, max_depth=depth)
                forest.fit(all_data[dat], Ytrain)
                RMSE_val_ridge = mean_squared_error(Yval, forest.predict(all_val[dat]), squared = False)
                
                if RMSE_val_ridge < best_rmse:
                    best_rmse = RMSE_val_ridge
                    estimators = est
                    data_portion = samples
                    best_depth = depth
                    best_data = dat
    print(est)
````

### KNN

Jako poslední zkusim knn model kde budu zkoumat vliv hyperparametrů počet sousedů, metrika (jen eukleidovská geometrie), vliv vzdálenosti sousedů od zkoumaného bodu a škálování.

````python
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html
# data std, minmax a default

best_rmse = 1000000
# (0-no scaler, 1-minmax, 2-standard)
best_data = 5
best_neighbours = 0
best_weights = None # uniform / distance
best_metrics = 1

all_data = [current_data_train, X_train_minmax, X_train_std]
all_val = [current_data_val, X_val_minmax, X_val_std]


for neigh in range(1, 50):
    for met in range(1, 3): # might try more
        for weight in ["uniform", "distance"]:
            knn = KNeighborsRegressor(n_neighbors=neigh, p=met, weights=weight)
            for i in range(3):
                knn.fit(all_data[i], Ytrain)
                RMSE_val_ridge = mean_squared_error(Yval, knn.predict(all_val[i]), squared = False)

                if RMSE_val_ridge < best_rmse:
                    best_data = i
                    best_neighbours = neigh
                    best_weights = weight
                    best_metrics = met
                    best_rmse = RMSE_val_ridge
            
````
